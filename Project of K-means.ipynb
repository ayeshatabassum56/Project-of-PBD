```python
#%pip install nltk
import nltk 
from nltk.corpus import stopwords
nltk.download ( 'stopwords') 
nltk.download ( 'punkt') 
from nltk.tokenize import word_tokenize 
text= word_tokenize (listToStr) 
data = [word for word in text if not word in stopwords . words ()]
print(data)
```

```python
data
```

```python
#import nltk
#nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 
lemma= WordNetLemmatizer() 
lem=[]
for r in data:
       lem.append(lemma.lemmatize (r))
```

```python
print (lemma.lemmatize( 'Coverage'))
```

```python
import re 
dataupdate= [ ] 
dataupdate=[re. sub('[^a-zA-z0-9]+ ', '', _) for _ in lem] 
```

```python
dataupdate
```

```python
# Now we do clustring
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.cluster import KMeans
from sklearn. metrics import adjusted_rand_score
#import warnings filter 
from warnings import simplefilter
```

```python
#ignore all future warnings
simplefilter(action ='ignore',category=FutureWarning)

vectorizer = TfidfVectorizer(stop_words= 'english')
X = vectorizer.fit_transform(dataupdate)
```

```python
true_k = 5
model = KMeans (n_clusters=true_k, init= 'k-means++', max_iter=100, n_init=1)
model.fit (X)
print ("Our cluster organized is : ") 
print (" ")

order = model.cluster_centers_.argsort()[:,::-1] 
terms = vectorizer.get_feature_names()
for i in range(true_k): 
        print(" ")
        print("Cluster %d:" % i), 
        for i in order[i, :10]:
                print( '%s' % terms[i])
                print(" ") 
                
```

```python
print("Search the word :")
Y = vectorizer.transform( ["elements"])
guessing= model.predict (Y)
print (guessing)

Y =vectorizer.transform ( ["claim that this cetral aspect of anarchism is definitive is to sell anarchism short"])
guessing = model.predict (Y) 
print (guessing)
```
